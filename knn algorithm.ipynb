```python
import pandas as pd
import numpy as np
 
import os
print(r"C:\Users\ASUS\Desktop\ml\KNN algrothim")
```

```python
df = pd.read_csv(r"C:\Users\ASUS\Desktop\ml\KNN algrothim\Placement_Data_Full_Class.csv")

# Select specific columns/features
features = ["hsc_p", "ssc_p", "salary"]
x = df[features]
```

```python
x.head()
```

```python
from sklearn.impute import SimpleImputer

my_imputer = SimpleImputer()
X = pd.DataFrame(my_imputer.fit_transform(x))
X.columns =x.columns
y= df["gender"]
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1, metric='minkowski',weights='distance')
knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)
```

```python
df= pd. DataFrame ( { 'Actual' : y_test, 'predicted':y_pred})
df
```

```python
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(y_test, y_pred)
print(cm)
```

```python
from sklearn.model_selection import cross_val_score
success = cross_val_score(estimator = knn, X=X_train, y=y_train , cv = 4)
print(success.mean())
print(success.std())
```

```python
output = pd.DataFrame({'sscp':X_test.ssc_p,'hscp':X_test.hsc_p,'Salary': X_test.salary,'Gender':y_test,'PredictGender':y_pred})
output.to_csv('KnnSubmission.csv', index=False)
print("Your submission was successfully saved!")
```
